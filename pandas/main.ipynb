## Series and Dataframes

```python
# 2 main datatypes

series = pd.Series(["BMW", "Honda", "Ford"])
print(series)
```

```python
# series = 1-dimentional
# created out of lists

colours = pd.Series(["Red", "Green", "Blue"])
print(colours)
```

```python
# dataframe = 2-dimentional
# created out of dicts

dataframe = pd.DataFrame({"model": series, "colour": colours})
print(dataframe)
```
